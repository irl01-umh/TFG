```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
url2="https://raw.githubusercontent.com/irl01-umh/TFG/main/train_proteins.csv"
url3="https://raw.githubusercontent.com/irl01-umh/TFG/main/train_clinical_data.csv"
#Carga de train_proteins.csv y train_clinical_data.csv

train_proteins = pd.read_csv(url2)
train_proteins.head(25)

train_clinical_data = pd.read_csv(url3)
train_clinical_data.head(25)
```

```python
columnas_comunes = ['visit_id', 'patient_id', 'visit_month']

datos_combinados = pd.merge(train_proteins, train_clinical_data, on=columnas_comunes)
datos_combinados

#convertir en excel
```

```python
df=pd.DataFrame(datos_combinados)

df.to_csv('datos_combinados_proteins_clinical.csv', index=False)
```
